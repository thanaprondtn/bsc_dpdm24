```python
# prompt: open sheet number 3 of the excel file and select only the table part

import pandas as pd

# Replace with the actual path to your Excel file
excel_file_path = '/content/drive/MyDrive/bsc_dpdm24_data570-8/export62_64.xlsx'

try:
  # Read sheet number 3 (index 2)
  df_excel = pd.read_excel(excel_file_path, sheet_name=2, header=None) # Assuming the table starts from the first row

  # Find the header row (assuming it's the first row with non-NaN values in the first column)
  header_row = 0
  for i in range(len(df_excel)):
      if not pd.isna(df_excel.iloc[i, 0]):
          header_row = i
          break

  # Select data from header row onwards
  df_table = df_excel.iloc[header_row:]

  # Set the header row as column names
  df_table.columns = df_table.iloc[0]

  # Remove the header row itself
  df_table = df_table.iloc[1:]

  # Reset index
  df_table = df_table.reset_index(drop = True)

  print(df_table.head())


except FileNotFoundError:
  print(f"Error: File not found at {excel_file_path}")
except ValueError as e:
  print(f"Error: {e}")  # More specific error message
except Exception as e:
  print(f"An unexpected error occurred: {e}")
```
